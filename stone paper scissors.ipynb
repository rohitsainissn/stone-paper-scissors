```python
from IPython.display import Image
import pyttsx3
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)
rate = engine.getProperty('rate')
engine.setProperty('rate', 150)
print('enter \n s for Stone \n c for Scissor\n p for Paper\n')
import random
lis = ["Stone","Scissors","Paper"]
def comp_chance(lis):
    pc=random.choice(lis)
    if pc =='Stone':
        display(Image(filename='stone.jpeg'))
    elif pc == 'Paper':
        display(Image(filename='paper.jpeg'))
    elif pc == 'Scissors':
        display(Image(filename='Scissors.jpeg'))
    return pc

r=c=0
name = input("enter your name \n")
def rohit_chance(rohit):
    if rohit == "s":
        return "Stone"
    elif rohit == "c":
        return "Scissors"
    elif rohit == "p":
        return "Paper"
    else:
        print ("wrong entry")
        engine.say("wrong entry")
        engine.runAndWait()
def w():
    global r
    r=r+1
    engine.say("congratulation ")

    return "congratulation"
def l():
    global c
    c=c+1
    engine.say('looser')

    return'looser'


wi =0
lo = 0
while r<5 and c <5 and wi<5 and lo<5:
    rohit = input("")
    if rohit =='s':
        display(Image(filename='stone.jpeg'))
        print('V/S')
    elif rohit == 'p':
        display(Image(filename='paper.jpeg'))
        print('V/S')
    elif rohit == 'c':
        display(Image(filename='Scissors.jpeg'))
        print('V/S')
    var = comp_chance(lis)

    engine.say(var)

    bari = print(f'{name} {rohit_chance(rohit)} V/S Computer {var}')

    if rohit_chance(rohit) == var:
        print ("this point is draw")
        engine.say('Tie')       

    if rohit_chance(rohit) == "Stone":
        if var=="Scissors":
            print (name,w())
            wi+=1
        elif var=="Paper":
            print (name,l())
            lo+=1


    elif rohit_chance(rohit) == "Scissors":
        if var=="Stone":
            print (name, l())
            lo+=1
        elif var=="Paper":
            print (name, w())
            wi+=1

    elif rohit_chance(rohit) == "Paper":
        if var=="Stone":
            print (name, w())
            wi+=1
        elif var=="Scissors":
            print (name, l())
            lo+=1
    print("Score", name, "V/S Computer", r, ":", c)
    sco1,sco2= str(r),str(c)
    engine.say(sco1)
    engine.say(sco2)
    engine.runAndWait()

print("End of the game")
engine.say("End of the game")

if r>c:
    print(name," won the game")
    engine.say("you win hurraaaa")
    engine.runAndWait()
    
elif c>r:
    print("Computer win, better luck next time")
    engine.say("Computer win, better luck next time")
    engine.runAndWait()
elif c==r:
    print("Match Draw")
    engine.say("Match Draw")
    engine.runAndWait()
    
final =f'final score", {name}, "V/S Computer",{r},":",{c}'
print(f'final score:- {name} V/S Computer {r} : {c}') 
engine.say(final)
engine.runAndWait()

```
